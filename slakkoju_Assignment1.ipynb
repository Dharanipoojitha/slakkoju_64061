```python
# Collection of IMDB
#Working with the IMDB dataset will include examining a collection of 50,000
from tensorflow.keras.datasets import imdb
(train_data, train_labels), (test_data, test_labels) = imdb.load_data(
    num_words=10000)
```

```python
#Decoding reviews go to the next
word_index = imdb.get_word_index()
reverse_word_index = dict(
    [(value, key) for (key, value) in word_index.items()])
decoded_review = " ".join(
    [reverse_word_index.get(i - 3, "?") for i in train_data[0]])
```

```python
decoded_review
```

```python
#Only the top 10,000 most common terms in the training set will be retained accordingly
#The list of reviews that make up the variable strain data and test data are listed
#As a result of our restriction to the top 10,000 most frequently occurring words

#Preparation of the data
#List of integers cannot be found into neural network. Our lists must be transformed.
# One possible solution is to create an integer tensor of the form samples
#To convert our list into vector of zeros and ones we could one hot and cut them.
#We will choose the second option now let’s activate our data which will be done.

import numpy as np
def vectorize_sequences(sequences, dimension=10000):
    results = np.zeros((len(sequences), dimension))
    for i, sequence in enumerate(sequences):
        results[i,sequence] = 1
    return results
# Our vectorized training data
x_train = vectorize_sequences(train_data)
# Our vectorized test data
x_test = vectorize_sequences(test_data)
```

```python
#Currently this is how our samples appear
x_train[0]
```

```python
x_test[0]
```

```python
#Vectoring our labels is something else we ought to do
y_train = np.asarray(train_labels).astype("float32")
y_test = np.asarray(test_labels).astype("float32")
```

```python
#A Neural network is now prepared to recieve our data
#Expanding our network
from tensorflow import keras
from tensorflow.keras import layers
model = keras.Sequential([
layers.Dense(16, activation="relu"),
layers.Dense(16, activation="relu"),
layers.Dense(1, activation="sigmoid")
])
```

```python
model.compile(optimizer="rmsprop",
loss="binary_crossentropy",
metrics=["accuracy"])
```

```python
#Verifying our methodology
#We will separate 10,000 samples from the original training data to establish validation
x_val = x_train[:10000]
partial_x_train = x_train[10000:]

y_val = y_train[:10000]
partial_y_train = y_train[10000:]
```

```python
#We will now train our model in mini batches of 512 samples for 20 each (20 iterations)
import tensorflow as tf # Import tensorflow
tf.random.set_seed(1234)
history = model.fit(partial_x_train,
                    partial_y_train,
                    epochs=20,
                    batch_size=512,
                    validation_data=(x_val, y_val))
```
